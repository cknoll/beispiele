```python
import sympy as sp
import symb_tools as st
import scipy as sc
import scipy.integrate
import numpy as np
import pylab as pl
%matplotlib inline
from IPython.display import display
from sympy.interactive import printing
printing.init_printing()
%reload_ext displaytools
```

```python
t, x1, x2, tau1 = sp.symbols('t, x1, x2, tau1')
xx = sp.Matrix([x1, x2])
```

Das System $\dot x = f(x) + G(x)\tau$ wird durch folgende Vektorfelder beschrieben:

```python
f = sp.Matrix([x2, -1])##
G = sp.Matrix([0, 1])##
```

```python
mod = st.SimulationModel(f, G, xx)
```

```python
x0 = [0.5, 0] # Anfangswert
T = 5.0
```

```python
tt = np.linspace(0, T, 1e3) # array mit Auswertezeitpunkten
```

```python
des_input = st.piece_wise((1, t <= 1 ), (t, t < 2), (0.5, t < 4), (1, True))
des_input_func = st.expr_to_func(t, des_input)
```

```python
rhs = mod.create_simfunction(input_function=des_input_func)
```

```python
XX = sc.integrate.odeint(rhs, x0, tt) # Durchfürhung der Simulation
ttau = des_input_func(tt)
```

```python
pl.figure(figsize=(10,5))
pl.subplot(2,2,1)
pl.plot(tt, XX, 'k-')
pl.subplot(2,2,2)
pl.plot(tt, ttau, 'k-')
```

Jetzt $\Psi$ und $\stackrel{\leftrightarrow}{I}_T$ festlegen:

```python
def Psi_x(xx):
    res = xx*1
    res[:, 0]*=-1
    return res

def Psi_tau(tau):
    return -1*tau

def time_inv(arr):
    if arr.ndim == 1:
        return 1*arr[::-1]
    else:
        assert arr.ndim ==2
        return 1*arr[::-1, :]    
```

```python
XX_bar = time_inv(Psi_x(XX))
ttau_bar = time_inv(Psi_tau(ttau))
```

```python
pl.figure(figsize=(10,8))
pl.subplot(3,2,1)
pl.plot(tt, XX, 'k-')
pl.subplot(3,2,2)
pl.plot(tt, ttau, 'k-')
pl.grid(1)
pl.subplot(3,2,3)
pl.plot(tt, Psi_x(XX), 'k-')
pl.subplot(3,2,4)
pl.plot(tt, Psi_tau(ttau), 'k-')
pl.grid(1)
pl.subplot(3,2,5)
pl.plot(tt, XX_bar, 'k-')
pl.subplot(3,2,6)
pl.plot(tt, ttau_bar, 'k-')
pl.grid(1)
```

Nachweis, dass $\bar{\mathbf x}(\cdot)$ und $\bar{\boldsymbol\tau}(\cdot)$ die Systemdynamik erfüllen:

```python
res = mod.num_trajectory_compatibility_test(tt, XX_bar, ttau_bar, full_output=True)
```
